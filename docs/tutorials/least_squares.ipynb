In this demo, we'll look at performing univariate least-squares regression using `bayesregress.fit_data`. We'll find a good fit to a complicated function with no user supplied information about what the function should be like, and without fine-tuning any parameters to get a good fit.

First we import:

```python
import numpy as np
from bayesregress import fit_data
import matplotlib.pyplot as plt
```

We set a seed so that this demo is reproducible:

```python
rng = np.random.default_rng(seed=1516)
```

Now, we generate some data:

```python
x = np.linspace(-10, 10, 901)
noise = rng.standard_normal(x.shape)
correct = 0.5 * x * np.cos(0.5 * x)
y = correct + noise
```

Then, we fit the data

```python
result = fit_data(x, y)
```

And we find the most likely model prediction:

```python
prediction = result.predict_for_map_model(x)
```

Plotting the data, the correct model, and the prediction shows that the fit is great, despite it being a complicated function:

```python
plt.plot(x, y, '.', label="Raw Data", ms=2)
plt.plot(x, correct, label='Correct Model', lw=2)
plt.plot(x, prediction, label='Inferred Model', lw=2)
plt.legend(loc='lower left')
```
